```python
import pandas as pd
import numpy as np
import folium
import matplotlib.pyplot as plt
import matplotlib
import base64
import seaborn as sns
import geojson
import geopandas

from folium.plugins import HeatMap
from datetime import datetime as dt
from io import BytesIO
from matplotlib.figure import Figure
from urllib.request import urlopen
from folium.features import GeoJsonPopup, GeoJsonTooltip

pd.plotting.register_matplotlib_converters()

%matplotlib inline
```

```python
crime = pd.read_csv("../Data/CleanedCrimeData-2021.csv")
```

```python
crime['OccurTime'] = crime['OccurTime'].apply(str)
crime['OccurTime'] = crime['OccurTime'].apply(lambda x: x.zfill(4))
crime['CrimeDate'] = crime.OccurDate + ' ' + crime.OccurTime
crime['CrimeDate'] = pd.to_datetime(crime['CrimeDate'])
cols = list(crime)
crime = crime.loc[:, cols]
crime.set_index('CrimeDate', inplace=True)
```

```python
with open("../GeoJson/Neighborhood_Boundaries.geojson") as f:
    hoods = geojson.load(f)
```

```python
hoods = geopandas.GeoDataFrame.from_features(hoods, crs="EPSG:4326")
```

```python
crimes = crime.groupby("Neighborhood").size()
```

```python
hoodmerge = hoods.merge(crimes.rename("crimes"), how="left", left_on ="MAPLABEL", right_on="Neighborhood")
```

```python
hoodmerge = hoods.merge(crime, how="inner", left_on ="MAPLABEL", right_on="Neighborhood")
```

```python
m = folium.Map(location=[45.4962, -122.4772], zoom_start=10)
```

```python
for _, r in hoods.iterrows():
    # Without simplifying the representation of each borough,
    # the map might not be displayed
    sim_geo = geopandas.GeoSeries(r['geometry']).simplify(tolerance=0.001)
    geo_j = sim_geo.to_json()
    geo_j = folium.GeoJson(data=geo_j,
                           style_function=lambda x: {'fillColor': 'orange'})
    folium.Popup(r['MAPLABEL']).add_to(geo_j)
    geo_j.add_to(m)
```
