## Creating a basic heatmap using folium. I adopted liberally from https://alysivji.github.io/getting-started-with-folium.html

```python
import folium
from folium import plugins
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
#I used official crime stats from the city of Portland: https://www.portlandoregon.gov/police/71978
#I created a filtered spreadsheet of just one kind of crime (weapons law violations), and made sure they all had lat/lon values,
#which did eliminate a few entries that were missing lat/lon
crime = pd.read_csv("../Data/CleanedCrimeData-2021.csv")
```

```python
crime['OccurTime'] = crime['OccurTime'].apply(str)
crime['OccurTime'] = crime['OccurTime'].apply(lambda x: x.zfill(4))
crime['CrimeDate'] = crime.OccurDate + ' ' + crime.OccurTime
crime['CrimeDate'] = pd.to_datetime(crime['CrimeDate'])
cols = list(crime)
crime = crime.loc[:, cols]
crime.set_index('CrimeDate', inplace=True)
```

```python
crime.head(1)
```

```python
m = folium.Map(location=[45.5291, -122.6768], zoom_start=12)
```

```python
for index, row in crime.iterrows():
    folium.CircleMarker([row['OpenDataLat'], row['OpenDataLon']],
                        radius=1,
                        popup=str(row['OccurDate']),
                        fill_color="#3db7e4", 
                       ).add_to(m)
```

```python
crimeArr = crime[['OpenDataLat', 'OpenDataLon']]
m.add_child(plugins.HeatMap(crimeArr, radius=0))
m
```
